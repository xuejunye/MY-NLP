```python
# -*- encoding:utf-8 -*-
import jieba.analyse
from os import path
from scipy.misc import imread
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

```

```python
d = path.dirname('.')
content = open(path.join(d, 'testing1.txt')).read()
   

# tags extraction based on TF-IDF algorithm提取关键词十个
tags = jieba.analyse.extract_tags(content, topK=20, withWeight=False)
text =" ".join(tags) 
print(text)
#是一篇专业论文，提取关键字看看
```

```python
# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:

plt.imshow(wordcloud)
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
    
 
```

```python
#词云配置，设置中文字体
wordcloud = WordCloud(font_path='zxq.ttf',max_font_size=40).generate(text)
```

```python
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
# 保存图片
wordcloud.to_file(path.join(d, "wordcloud.png"))
```

```python
from PIL import Image

# 打开一个jpg图像文件，注意是当前路径:
im = Image.open('background.jpg')
# 获得图像尺寸:
w, h = im.size
print('Original image size: %sx%s' % (w, h))
# 缩放到50%:
im.thumbnail((w//2, h//2))
print('Resize image to: %sx%s' % (w//2, h//2))
plt.figure()
plt.imshow(im)
plt.axis("off")
plt.show()
```

```python
    from scipy.misc import imread
    from PIL import Image
    import matplotlib.pyplot as plt 
    from wordcloud import WordCloud
    texts = open('testing1.txt').read()
    w = imread('background.jpg')
# read the mask词云的背景设置（配置词云参数）
    wc = WordCloud(font_path='zxq.ttf', # 设置字体
                          background_color="white", #背景颜色
                          max_words=120, #词云显示的最大词数
                          mask=w, #设置背景图片
                          max_font_size=120, #字体最大值 #      random_state=42,
                         )
    wordcloud = wc.generate(texts)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```
